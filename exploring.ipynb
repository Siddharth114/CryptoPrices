```python
reg_slopes = dict()
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(binancecoin_data['SNo'], binancecoin_data['Close'])
ax = sns.regplot(x='SNo', y='Close', data=binancecoin_data, scatter_kws = {'color': 'g',  'alpha':0.2}, line_kws={'lw':4, 'label':"y={0:.3f}x+{1:.1f}".format(slope,intercept)})
ax.legend()
reg_slopes['binancecoin']=slope
```

```python
reg_slopes = {k: v for k, v in sorted(reg_slopes.items(), key=lambda item: item[1], reverse=True)}
reg_slopes
```

From the slopes of the regression lines, we see that the fastest growing cryptocurrencies are bitcoin, binance coin, and ethereum.

Using a Decision Tree Regressor model to predict the closing price of the different cryptos on a certain date, along with the accuracy of the model.

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn import metrics
from sklearn.model_selection import train_test_split 
X = binancecoin_data[['SNo']]
Y = binancecoin_data[['Close']]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=.20, random_state=0)
DT_model = DecisionTreeRegressor(max_depth=5).fit(X_train,Y_train)
DT_predict = DT_model.predict(X_test)
print("Accuracy:",DecisionTreeRegressor.score(DT_model, X_test, Y_test))
```
