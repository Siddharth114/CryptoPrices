```python
import pandas as pd
import seaborn as sns
from scipy import stats
import matplotlib.pylab as plt
```

```python
bitcoin_data = pd.read_csv('Data/coin_Bitcoin.csv')
bitcoin_data.head()
```

```python
cardano_data = pd.read_csv('Data/coin_Cardano.csv')
cardano_data.head()
```

```python
dogecoin_data = pd.read_csv('Data/coin_Dogecoin.csv')
dogecoin_data.head()
```

```python
ethereum_data = pd.read_csv('Data/coin_Ethereum.csv')
ethereum_data.head()
```

```python
litecoin_data = pd.read_csv('Data/coin_Litecoin.csv')
litecoin_data.head()
```

```python
monero_data = pd.read_csv('Data/coin_Monero.csv')
monero_data.head()
```

```python
xrp_data = pd.read_csv('Data/coin_XRP.csv')
xrp_data.head()
```

```python
reg_slopes = dict()
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(bitcoin_data['SNo'], bitcoin_data['Close'])
ax = sns.regplot(x='SNo', y='Close', data=bitcoin_data, scatter_kws = {'color': 'g',  'alpha':0.2}, line_kws={'lw':4, 'label':"y={0:.2f}x+{1:.1f}".format(slope,intercept)})
ax.legend()
reg_slopes['bitcoin']=slope
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(cardano_data['SNo'], cardano_data['Close'])
ax = sns.regplot(x='SNo', y='Close', data=cardano_data, scatter_kws = {'color': 'g',  'alpha':0.2}, line_kws={'lw':4, 'label':"y={0:.6f}x+{1:.1f}".format(slope,intercept)})
ax.legend()
reg_slopes['cardano']=slope
```
